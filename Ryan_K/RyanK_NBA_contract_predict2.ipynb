```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#import data and view data frame
data = pd.read_csv(
    "../Resources/freeagent_2024_salary.csv")

data = pd.DataFrame(data)

data.columns

```

```python
# Drop unnecessary columns
data = data.drop(columns=['Player_Name', 'PLAYER_ID','TEAM_ABBREVIATION','LEAGUE_ID','TEAM_ID'])
data
```

```python
# Create scaler instance
scaler = StandardScaler()

#fit standardscaler
data_scaler = scaler.fit(data)

#scale the data
scaled_data = data_scaler.transform(data)
```

```python
# separate preprocessed data into features and target arrays
y = data['1819_salary_adjusted'].values
x = data.drop(columns='1819_salary_adjusted').values


# split preprocessed data into trainin and testing data
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=8)
```

```python
# create linear regression model
# 

reg = SGDRegressor()
reg.fit(X_train, y_train)
```

```python
error = y_test - reg.predict(X_test)
error 
```

```python
y_pred = reg.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # setting squared=False gives RMSE directly
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```
