```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import playercareerstats
import pandas as pd  

def getIDs(names):
    nba_players = players.get_players()
    id_mapping = {}

    for name in names:
        for player in nba_players:
            if player['full_name'] == name:
                id_mapping[name] = player['id']
                break
        else:
            id_mapping[name] = None
    
    return id_mapping

def getCareerStats(player_ids):
    career_stats = {}

    for name, player_id in player_ids.items():
        if player_id:
            career = playercareerstats.PlayerCareerStats(player_id=player_id)
            career_data = career.get_data_frames()[0]
            career_stats[name] = career_data
        else:
            career_stats[name] = None
    
    return career_stats

names = ['Kareem Abdul-Jabbar', 'LeBron James', 'Michael Jordan']
id_mapping = getIDs(names)
career_stats = getCareerStats(id_mapping)

# Print career stats as pandas DataFrames
for name, stats in career_stats.items():
    print(f"Career Stats for {name}:\n")
    if stats is not None:
        print(stats)
        print("\n")
    else:
        print("Player not found or no career stats available.\n")



```

```python
from nba_api.stats.endpoints import playerdashboardbyyearoveryear
from nba_api.stats.static import players
import pandas as pd

# Read the CSV file into a DataFrame
data = pd.read_csv("nba_contracts_history.csv")

# Extract values from a specific column and convert them to a list
player_names = data["NAME"].tolist()

def getIDs(names):
    nba_players = players.get_players()
    id_mapping = {}

    for name in names:
        for player in nba_players:
            if player['full_name'] == name:
                id_mapping[name] = player['id']
                break
        else:
            id_mapping[name] = None
    
    return id_mapping

# List of player full names
# player_names = ["LeBron James", "Kevin Durant", "Kobe Bryant"]

# Get player IDs using the getIDs function
player_ids = getIDs(player_names)

# Year 2010-2011
season = "2010-11"

# Initialize a list to store player statistics
player_stats = []

# Loop through each player and retrieve statistics
for name, player_id in player_ids.items():
    if player_id is not None:
        response = playerdashboardbyyearoveryear.PlayerDashboardByYearOverYear(
            player_id=player_id,
            season=season
        )
        
        # Get the row with player statistics
        player_row = response.get_dict()['resultSets'][1]['rowSet'][0]
        
        # Extract PTS, REB, AST, and GP
        pts = player_row[29]  # Total points
        reb = player_row[20]  # Total rebounds
        ast = player_row[22]  # Total assists
        gp = player_row[5]   # Total games played
        
        player_stats.append([name, pts, reb, ast, gp])
    else:
        print(f"No player found with the name: {name}")

# Create a pandas DataFrame from the player statistics
columns = ["Name", "PTS", "REB", "AST", "GP"]
player_stats_df = pd.DataFrame(player_stats, columns=columns)

# Export the DataFrame to a CSV file
player_stats_df.to_csv("player_stats_2010_2011.csv", index=False)

print("Player statistics exported to player_stats_2010_2011.csv")


```
